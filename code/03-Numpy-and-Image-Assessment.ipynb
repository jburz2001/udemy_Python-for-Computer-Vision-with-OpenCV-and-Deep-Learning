**TASK: Run the cell below to create an array of random numbers and see if you can figure out how it works.**

```python
# This line sets a "seed" so you get the same random numbers we do
np.random.seed(101)
# This line creates an array of random numbers
arr = np.random.randint(low=0,high=100,size=(5,5))
```

**TASK: What are the largest and smalled values in this array?**

The smallest value is 0, inclusive, while the largest value is 100, exclusive.
In terms of oversimplification, 0 can be reached as a minimum value, but 100 cannot be reached as a maximum.
In terms of math, range = [0,100)


```python

min,max = np.min(arr),np.max(arr)
print("The max value of arr: {}\n The min value of arr: {}".format(min,max))
```

**TASK: Use PIL and matplotlib to read and display the ../DATA/00-puppy.jpg image.**

```python

import matplotlib.pyplot as plt
from PIL import Image

#read file from relative path
relPath = r"..\DATA\00-puppy.jpg"
img = Image.open(relPath)

#display image
plt.imshow(img)
```

**TASK: Convert the image to a NumPy Array**

```python
npImg = np.asarray(img)
```

**FINAL TASK: Use slicing to set the RED and GREEN channels of the picture to 0, then use imshow() to show the isolated blue channel**

```python
#make copy of image
alteredImg = npImg.copy()

#iterate through 3rd axis
for i in [0,1]:
    #set R and G to 0
    alteredImg[:,:,i] = 0

plt.imshow(alteredImg)
```
